# Notebook introduction

Living labs describe the paradigm of working with new ideas and technology, and directly engaging with and observing users while they are living their lives.

While the levels of direct user engagement and co-creation vary between the examples referenced in this section, they share the access to users’ behavior in response to novel products, content, or activity.


# 1. Living labs

A living lab can be established using existing infrastructure and data sources. This was shown in the example of the Andorra living lab (discussed in the video content) which made use of CDRs, credit card transactions, and public transportation data. While direct user engagement may be limited, their behaviors – especially their reactions to new products or experimental interventions – can be observed in this setting. Living labs can also be entirely virtual, as is the case in the A/B testing of web applications.


Although the bleeding edge of living lab development is driven by the growing data collection and interaction capabilities that have been enabled by the spread of ubiquitous computing, this is not a prerequisite for big or electronic data. In the video content, Professor Alex Pentland discusses an example of how Nike leverages living labs to determine which shoes to release within their stores. #### a) How do "deep data" and "living labs" help in avoiding the pitfalls of only studying available data sets?

With “deep data”, such as provided by “living labs”, we avoid the following pitfalls of studying available data sets:

>* The considerable overheads in processing poor quality available big data - as we need to do a lot of filtering, data slicing and floundering around before we get to the meaningful data we need, which may also turn out to be of poor quality.


>* Being distracted over spending too much time studying the data - instead of studying the interactions and behavior of the population.



#### b) List some of the advantages of "living labs" (as opposed to "focus groups" or "large scale surveys").

Advantages of “living labs” include:

>*	Allowing us to be more focused on the population we want to study, without the limitations of a big data mirage. 


>*	Providing us the richness of high resolution data across all the network types underlying social behaviors, so we can ask “deeper” questions on social behavior and validate the consistency of our hypothesis across multiple network types.


>*	Allowing the test population to live in a natural way since the data collection method is non-obtrusive and non-disruptive of their day-to-day activities.


>*	Facilitating the instrumentation of behavior change and being able to test hypothesis more accurately.


>*	More access to ground truth validation, since the test population is selected by design, instead of being randomly sampled.

In the video content, Professor Alex Pentland discussed how many organizations have begun to realize the value of data as a strategic asset. Keeping this in mind, consider the following question: 

- How does the role of an analyst change when considering strategic analysis as opposed to more traditional data analysis?

**Hint**: 
> Provide a short description of the typical tasks that you would expect within this new role, and briefly discuss or refer to potential organizational changes or parties that the analysts would need to implement or interact with in their new role.

Your markdown answer here.


When considering strategic analysis as opposed to more traditional data analysis, the role of an analyst dramatically changes in the following ways:

>*	The strategic analyst will no longer work in the silo of a business unit or project. He/she will step up into a key role across the enterprise, with an end to end view of the impact of big data and analytics on corporate business outcomes.


>*	The strategic analyst will understand corporate business drivers and will continuously look for opportunities in big data to increase the company’s revenue and competitive edge.


>*	The strategic analyst will test hypothesis and be accountable for interpretation, recommendations, visualization and story-telling of the findings of big data analytics.


>*	To fulfil this new role, the strategic analyst will need to be in a senior position in the company org chart, so as to be empowered to make decisions as well as being able to engage at the right level with other senior managers to collaborate on their exploitation of big data and social analytics.


>*	The strategic analyst will need to interact directly with business leaders to understand their requirements and aspirations for big data, even to the point of prompting and educating business leaders who are not already exposed to big data opportunities for innovation and beating the competition.


>*	The strategic analyst will interact with data analysts, IT and project management resources to provide consultation and guidance on achieving meaningful big data results for the business.
